```python
class SimpleReflexDoorAgent:
    def decide(self, sensor_input):
        """
        sensor_input: "Person" or "NoPerson"
        returns: "Open" or "Close"
        """
        if sensor_input == "Person":
            return "Open"
        else:
            return "Close"
```

```python
class DoorEnvironment:
    def __init__(self):
        self.door_state = "Closed"   # "Open" or "Closed"
        self.sensor = "NoPerson"     # "Person" or "NoPerson"

    def update(self, action):
        if action == "Open":
            self.door_state = "Open"
        else:
            self.door_state = "Closed"
```

```python
agent = SimpleReflexDoorAgent()
env = DoorEnvironment()

inputs = ["NoPerson", "Person", "Person", "NoPerson", "NoPerson"]

for step, sensor in enumerate(inputs):
    action = agent.decide(sensor)
    env.update(action)
    print(f"Step {step+1}: Sensor={sensor}, Action={action}, Door={env.door_state}")
```

```python
import matplotlib.pyplot as plt
from IPython.display import clear_output
import time
```

```python
def draw_door(state, step=None, sensor=None, action=None):
    """
    state: "Open" or "Closed"
    step: current step number (int)
    sensor: "Person" or "NoPerson"
    action: "Open" or "Close"
    """
    clear_output(wait=True)

    plt.figure(figsize=(4, 6))

    # Draw door frame
    # Frame is always same
    plt.plot([0, 1], [0, 0])   # bottom
    plt.plot([0, 1], [5, 5])   # top
    plt.plot([0, 0], [0, 5])   # left
    plt.plot([1, 1], [0, 5])   # right

    if state == "Closed":
        # Filled door (closed)
        plt.fill_between([0, 1], 0, 5, alpha=0.5)
        title = "Door: CLOSED"
    else:
        # Open door (just frame, gap inside)
        # Draw a smaller panel to look like open
        plt.fill_between([0.7, 1], 0, 5, alpha=0.5)
        title = "Door: OPEN"

    if step is not None:
        title = f"Step {step} | {title}"
    if sensor is not None and action is not None:
        title += f"\nSensor: {sensor} | Action: {action}"

    plt.title(title)
    plt.xlim(-0.5, 1.5)
    plt.ylim(-0.5, 5.5)
    plt.axis('off')
    plt.show()
```

```python
draw_door("Closed", step=0, sensor="NoPerson", action="Close")
time.sleep(1)
draw_door("Open", step=0, sensor="Person", action="Open")
```

```python
def run_visual_door_simulation():
    agent = SimpleReflexDoorAgent()
    env = DoorEnvironment()

    # Sequence of sensor inputs over time
    # You can change / extend this list
    inputs = ["NoPerson", "Person", "Person", "NoPerson", "NoPerson", "Person", "NoPerson"]

    for step, sensor in enumerate(inputs, start=1):
        # perception
        env.sensor = sensor

        # agent decides
        action = agent.decide(env.sensor)

        # environment updates
        env.update(action)

        # draw visual door
        draw_door(env.door_state, step=step, sensor=sensor, action=action)

        # small pause so you can see animation
        time.sleep(0.8)

    print("Simulation finished.")
```

```python
run_visual_door_simulation()
```
